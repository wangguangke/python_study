# 3. 程序控制

一般来说，Python中的代码按照其出现的顺序逐行执行。程序代码从第一行开始，按照自上而下的顺序依次执行，执行完某一行后再执行下一行，这种顺序执行方式通常称为程序的顺序结构。

实际编程过程中，我们通常需要对某些条件进行判断，并根据判断的结果进行不同的处理。例如，判断一个学生的考试成绩是不是大于60分，如果大于60分则认为其考试合格，否则认为不合格。有时候也需要对某些操作循环执行多次。

因此，Python中的程序控制结构可以划分为选择结构和循环结构。

## 3.1 选择结构

计算机在处理问题的时候也需要做判断，根据判断结果控制程序的执行流程，这种程序执行方式被称为选择结构。


Python中的程序选择结构分为简单分支选择、双分支选择和多分支选择三种结构，通过关键字包括if、else和elif来实现选择。

### 3.1.1 条件测试

Python中程序选择结构的核心是一个值为True或者False的表达式，这种表达式称为条件测试。常见的条件测试可以划分为单条件测试和多条件测试两类。

1. 单条件测试。单条件测试是最简单的测试形式，用于检查变量的值是否与特定的值相等。Python中检查是否相等的符号是使用“==”，当特定的值与某变量名所指向的值相等时，结果为True，否则为False。检查的对象可以是字符串，也可以是数字

```python
car='bmw'
car=='bmw'
```

```python
# 在检查字符串是否相等时，需要区分字母的大小写。大小写不同的两个值是被视为不相等的
car='Bmw'
car=='bmw'
```

```python
# 如果字母的大小无关紧要，只想检查变量的值，可以将变量的值转换为小写后再进行比较，
car='Bmw'
car.lower()=='bmw'
```

除了使用判断两个值是否相等外，还可以判断两个值是否不相等，可以结合使用惊叹号和等号（！=）来判断，其中的惊叹号（！）表示不。当两个值不相等时返回True，相等是返回False。

```python
'Bmw'!='bmw'
```

当检查的对象是数字时，仍然是是用双等号（==）。除此之外，还可以使用各种数学比较符，如小于（<），小于等于（<=）,大于（>），大于等于（>=）。

```python
age=18
age==18
# age<=21
# age>21
```

2.	多条件测试。

在实际编程中通常会检查多个条件，例如需要在两个或多个条件都为True时才执行相应的操作，而有时候只要求一个条件为True。这种情况下可以使用and和or进行多条件测试。

 •	使用and检查多个条件: 要检查两个条件是否都为True，可以使用关键字and将两个条件测试合二为一。如果每个测试都为True，整个表达式就为True；如果至少一个测试没有通过，整个表达式就为False。

```python
age_0=20
age_1=17
age_0<18 and age_1<18
```

•	使用or检查多个条件:or用于检测多个条件中至少有一个条件通过测试，整个表达式就能通过。即，只要有一个条件测试结果为True，整个表达是的结果就为True；仅当所有检测条件都没有通过，结果都为False，整个表达是的结果才为False。

```python
# 要检查两个人的年龄至少有一个小于18岁
age_0=20
age_1=17
age_0<18 and age_1<18
```

### 3.1.2 if 语句

if语句是由一个if条件表达式实现，单个if语句又称为单分支if语句。单分支的if语句是最基本的选择语句，它根据一个布尔表达式的结果来选择是否执行一条语句。



如果条件测试的值为True，Python就执行紧跟在if语句后面的代码；如果为False，Python就忽略这些代码。

单个if语句的语法只有一个测试和一个操作，语法规则如下:

if 表达式:

    执行语句1
    
    执行语句2

注意：在Python中，if语句结束用冒号 “:” 表示，执行语句前要加四个空格，这是python中的缩进，具有共同缩进的语句，为同一个整体。也就是说，上面例子中的“执行语句1”和“执行语句2”是同一个整体。

```python
#例1 根据考试成绩，判断是否及格。
grad=50
if grad>=60:
    print('您的考核结果为：及格')
    print('祝贺你！')
```

### 3.1.3 if…else语句

在上面的if语句中，如果测试条件不为True的话，if后面的语句块将不再执行。这种结构是不友好的。If语句后可跟一个可选的else语句，供测试条件不通过时执行。该结构为if…else语句。

if 表达式:

    语句块1
    
else:

    语句块2




```python
# 根据输入的成绩，判断学生是否通过考试

grade=int(input('请输入成绩:'))
if grade <60:
    print('考试不及格！')
else:    
    print('恭喜您，考试及格！')
```

### 3.1.4 if…elif…else语句

我们经常需要检查超过两个条件的情形，当需要在一个条件为False后再检查其他条件。为此，我们可以使用Python提供的if-elif-else的结构。

Python只执行if-elif-else结构中的一个代码块，它依次检查每个条件测试，直到遇到通过了的条件测试。测试通过后，Python将执行紧跟在它后面的代码，并跳过余下的测试。if-elif-else语句的一般形式如下:

if 表达式1:

    语句块1
    
elif 表达式2:

    语句块2
    
else:

    语句块3



```python
#例：根据年龄段收取游乐场门票，规则如下：4岁以下免票，4-18岁收取25元，18岁及以上收取40元。
age=int(input('请输入年龄：'))
if age<4:
    print('票价0元')
elif age<18:
    print('票价25元')
    
elif age<60:
    print('票价40元')
else:
    print('票价10元')
```

除了使用简单的if-elif-else语句之外，还可以根据需要使用任意数量的elif代码块，即：一个嵌套的if结构中可以有一个if…elif…else嵌套在另一个if…elif…else结构中。

```python
if 表达式1:
    语句块1
elif 表达式2:
    语句块2
elif 表达式3:
    语句块3
else:
    语句块4
```

```python
# 在上例基础上增加一个检查，65岁及以上的老人可半价购买门票。

age=int(input('请输入年龄：'))
if age<4:
    print('票价0元')
elif age<18:
    print('票价25元')
elif age<65:
    print('票价40元')
else:
    print('票价20元')
```

### 3.2.2 for 循环

for循环，也称为迭代循环(迭代即重复相同的逻辑操作)，迭代循环是Python中最强大的循环结构。for循环可以遍历任何序列的项目，如一个列表、一个元组、一个集合、一个字典或一个字符串等迭代对象。

for循环语句总是包含以下几个关键部分：for关键字、变量名、in关键字、冒号和缩进的代码块。

一般来说，for循环语句可以分为单层循环和多重循环。

```python
# 字符打印
msg='hello'
for s in msg:
    print(s)
```

```python
# 配合range函数，多次打印人名
print('My name is:')
for i in range(5):
    print('张三 ({0})'.format(i))
```

```python
# 思考：用for循环计算0-100以内的偶数和

sum = 0
for i in range(2,101,2):
#     sum=0
    sum += i
print(sum)
```

多重循环：Python编程语言中的for循环还允许嵌套使用，即允许在一个循环中使用另外一个循环，语法结构如下：

```python
for 循环变量1 in 序列1:
    for 循环变量2 in 序列2:
        语句块
```

```python
#例：用for循环打印九九乘法表

for i in range(1,10):
    for j in range(1,i+1):
        print('{0}*{1}={2}'.format(i, j, i*j),end='\t')
    print('')
```

### 3.2.3 while循环

可以用while循环来做和for循环同样的事，只是for循环比while循环更简洁。While循环也被称为条件循环，语法结构如下：

while 条件表达式：

    语句块。
    
while循环语句总是包含以下几个部分：while关键字、条件表达式、冒号、缩进的代码块。

在while语句中，条件表达可以是任何表达式，True表示为任何非零值。

While语句后要紧跟着冒号“:”表示循环体的开始。

循环体中的语句可以是一个单的语句或一组具有统一缩进的语句块。当表达式的测试结果为True时执行语句块中的代码，当表达式的测试结果变为False时，程序控制跳出while循环体，继续执行while语句之后的代码。

```python
# 用while循环打印5次人名
print('My name is:')
i=0
while i<5:
    print('张三 ({0})'.format(i))
    i=i+1
```

```python
#用while循环求1~100的偶数和
sum=0
x=1
while x<=100:
    if x%2==0:
        sum+=x
    x=x+1
print(sum)
```

### 3.2.4 循环中的else语句

Python支持与循环语句相关联的else语句。如果else语句与for 循环一起使用， 则在循环遍历迭代对象结束后执行else语句；如果else语句while 循环一起使用的话，则在测试条件变为False时执行else中的语句。

```python
# for循环中的else
print('My name is:')
for i in range(5):
    print('张三 ({0})'.format(i))
else:
    print('打印完成')
```

```python
# while循环中的else
count=0
while count<5:
    print('{0}<5'.format(count))
    count+=1
else:
    print('{0}=5'.format(count))
```

### 3.2.5 跳出循环break和continue

在for和while循环体中，如果没有退出条件的话，循环会一直执行，直到条件为假或序列元素用完时才会结束，如果需要提前中断循环，比如循环结果已达预期，不想继续执行循环来占用资源，就可以使用break和continue这两条循环控制语句来跳出循环

1. 当执行break语句时，即使循环条件依旧为True，也会停止执行循环语句。

在代码中，break语句仅包含break关键字。break语句一般会结合if语句搭配使用，表示在某种条件下跳出循环体。如果break语句用在嵌套循环结构中，会停止执行break所在层的循环体，并继续执行外层循环结构的代码。



```python
for i in range(0, 10):
    print("i的值是: ", i)
    if i == 2:       
        break      # 执行该语句时将结束循环
```

```python
# 让用户不断的通过键盘输入内容，直到用户输入“quit”后退出循环
while True:
    
    content=input('请输入内容：')
    if content=='quit':
        break
    else:
        print('你输入的内容是:{0}'.format(content))
print('程序结束')
```

2. continue语句

与break语句一样， continue语句也是可以用在while循环和for循环内部，但不同的是，continue语句的作用是跳过当前循环的剩余语句，跳回循环开始处，重新对循环条件求值，即：执行下一次循环。



```python
# 输出1-10的奇数
current_number=0
while current_number<10:
    current_number+=1
    if current_number % 2==0:
        continue
    print(current_number)
```

```python
# 输出1-10之间的奇数之和
sum=0
current_number=0
while current_number<10:
    current_number+=1
    if current_number % 2==0:
        continue
    else:
        sum+=current_number
print("1-10以内的奇数和为:{0}".format(sum))

```

## 3.3 异常处理

异常是一个事件，这个事件会在程序执行过程中发生，影响程序的正常进行。

一般情况下，在Python程序无法正常执行时就会发生异常。

异常是Python的特殊对象，它表示一个错误，Python程序在发生异常时需要对异常进行捕获及处理，否则会被迫终止执行。

```python
# 例如，当一个数与0相除时，就会报错：
10/0
```

1. try...except

异常的处理通常使用try……except语句。其中try子句中的代码块包含着可能会引发异常的语句，而except子句则用来捕捉相应的异常。如果try子句中的代码引发异常并被except子句捕捉，则执行except子句的代码块；如果try子句中的代码块没有出现异常，则继续往下执行异常处理结构之后的代码；如果出现异常但没有被except捕获，则继续往外层抛出；如果所有层都没有捕获并处理该异常，则程序崩溃并将该异常呈现给最终用户。

```python
try:
    a=10
    b='3'
    a/b
    10/0
    
except:
    print('程序错误！')
```

```python
try:
    a=1/0
except ZeroDivisionError:
    print('程序报错')
```

2. try…except…else

还可以使用try…except…else语句.带有else子句的异常处理结构可以看作是一种特殊的选择结构。如果try子句中的代码抛出了异常并且被某个except语句捕捉则执行相应的异常处理代码，这种情况下就不会执行else中的代码；如果try子句中的代码没有抛出异常，则执行else子句中的代码块。

```python
try:
    a=3/0
    print("{0}/{1}={2}".format(3,2,a))
except ZeroDivisionError:
    print('程序报错')
else:
    print('顺利完成，没有错误')
```

3. 多个except异常处理结构

在实际开发中，同一段代码可能会抛出多种异常，并且需要针对不同的异常类型对代码进行相应的处理。为了支持多种异常的捕获和处理，Python提供了带有多个except的异常处理结构，一旦某个except子句捕捉到了异常，则其它的except子句将不会再尝试捕捉异常。

```python
a=10
b='3'
a/b
```

```python
try:
    #y1=10/0
#     a=10
#     b='3'
#     y2=a/b
    10/3
except ZeroDivisionError:
    print('0不能被作为除数')
except TypeError:
    print('数据类型错误')
else:
    print('程序正常运行')
```

```python
while True:
    try:
        x=float(input('请输入被除数:'))
        y=float(input('请输入除数:'))
        z=float(x/y)
    except ZeroDivisionError:
        print('除数不能为0！')
    except ValueError:
        print('被除数和除数都应该为数值！')
    else:
        print('{0}/{1}={2}'.format(x,y,z))
        break
```

4. try…except…finally语句

在此结构中，无论try中的代码是否会发生异常，也不管抛出的异常有没有被except语句捕捉到，finally子句中的代码总会得到执行。

因此，finally语句中的代码常用来做一些清理工作以释放try子句中申请的资源。如果try子句中的异常没有被except捕获和处理，或者except子句或else子句中的代码抛出了异常，那么这些异常将会在finally子句执行完毕后再次抛出。

```python
# 使用finally语句中实现对变量值的重置
while True:
    try:
        x=float(input('请输入被除数:'))
        y=float(input('请输入除数:'))
        z=float(x/y)
    except ZeroDivisionError:
        print('除数不能为0！')
    except ValueError:
        print('被除数和除数都应该为数值！')
    else:
        print('{0}/{1}={2}'.format(x,y,z))
    finally:
        z=0
    print('z的值为：{0}'.format(z))
```

### 练习

编写一个加法器，将计算结果打印出来。但是有时候用户提供的是文本而不是数字，在这种情况下，当你尝试将输入转换为整数时，将引发 TypeError异常。在用户输入的任何一个值不是数字时都捕获 TypeError 异常，并打印一条友好的错误消息。

```python
while True:
    num1=input('请输入第一个数:')
    num2=input('请输入第二个数:')
    try:
        res=int(num1)+int(num2)
    except TypeError:
        print('数据类型错误')
    else:
        print('{}+{}={}'.format(num1,num2,res))
        break
```
